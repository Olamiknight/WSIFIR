```python
#Metrics

#Compute mutual information between th rigid transformed axis-reg image and
mi1 = ants.image_mutual_information(reg_image_cropped_ants_1_padded, rigid_transformed_image_1)
mi2 = ants.image_mutual_information(reg_image_cropped_ants_2_padded, rigid_transformed_image_2)
mi3 = ants.image_mutual_information(reg_image_cropped_ants_3_padded, rigid_transformed_image_3)


#Compute normalized cross correlation between the fixed axis image and the rigid transformed axis-tbet image
cc1 = ants.math.image_similarity(reg_image_cropped_ants_1_padded, rigid_transformed_image_1, 'Correlation')
cc2 = ants.math.image_similarity(reg_image_cropped_ants_2_padded, rigid_transformed_image_2, 'Correlation')
cc3 = ants.math.image_similarity(reg_image_cropped_ants_3_padded, rigid_transformed_image_3, 'Correlation')

#Compute mean squares between the fixed axis image and the rigid transformed axis-tbet image
mse1 = ants.math.image_similarity(reg_image_cropped_ants_1_padded, rigid_transformed_image_1, 'MeanSquares')
mse2 = ants.math.image_similarity(reg_image_cropped_ants_2_padded, rigid_transformed_image_2, 'MeanSquares')
mse3 = ants.math.image_similarity(reg_image_cropped_ants_3_padded, rigid_transformed_image_3, 'MeanSquares')
```

```python
# Binarize the arrays (threshold = 0.5)
reg_image_1_binary = (reg_image_cropped_ants_1_padded.numpy() > 0.5).astype(int)
reg_image_2_binary = (reg_image_cropped_ants_2_padded.numpy() > 0.5).astype(int)
reg_image_3_binary = (reg_image_cropped_ants_3_padded.numpy() > 0.5).astype(int)

rigid_transformed_image_1_binary = (rigid_transformed_image_1.numpy() > 0.5).astype(int)
rigid_transformed_image_2_binary = (rigid_transformed_image_2.numpy() > 0.5).astype(int)
rigid_transformed_image_3_binary = (rigid_transformed_image_3.numpy() > 0.5).astype(int)



# Compute Jaccard Index

jaccard_1 = jaccard_score(reg_image_1_binary.flatten(), rigid_transformed_image_1_binary.flatten())
jaccard_2 = jaccard_score(reg_image_2_binary.flatten(), rigid_transformed_image_2_binary.flatten())
jaccard_3 = jaccard_score(reg_image_3_binary.flatten(), rigid_transformed_image_3_binary.flatten())


# Compute Dice Similarity Score

dice1 = f1_score(reg_image_1_binary.flatten(), rigid_transformed_image_1_binary.flatten())
dice2 = f1_score(reg_image_2_binary.flatten(), rigid_transformed_image_2_binary.flatten())
dice3 = f1_score(reg_image_3_binary.flatten(), rigid_transformed_image_3_binary.flatten())


# Compute Hausdorff Distance using SimpleITK
sitk_initial_image_1 = sitk.GetImageFromArray(reg_image_cropped_ants_1_padded.numpy())
sitk_final_image_1 = sitk.GetImageFromArray(rigid_transformed_image_1.numpy())

sitk_initial_image_2 = sitk.GetImageFromArray(reg_image_cropped_ants_2_padded.numpy())
sitk_final_image_2 = sitk.GetImageFromArray(rigid_transformed_image_2.numpy())

sitk_initial_image_3 = sitk.GetImageFromArray(reg_image_cropped_ants_3_padded.numpy())
sitk_final_image_3 = sitk.GetImageFromArray(rigid_transformed_image_3.numpy())



# Compute Hausdorff Distance for axis image 1 and final image 1
hausdorff_filter_1 = sitk.HausdorffDistanceImageFilter()
hausdorff_filter_1.Execute(sitk_initial_image_1, sitk_final_image_1)
hausdorff_sitk_1 = hausdorff_filter_1.GetHausdorffDistance()

#Compute Hausdorff Distance for axis image 2 and final image 2
hausdorff_filter_2 = sitk.HausdorffDistanceImageFilter()
hausdorff_filter_2.Execute(sitk_initial_image_2, sitk_final_image_2)
hausdorff_sitk_2 = hausdorff_filter_2.GetHausdorffDistance()

#Compute Hausdorff Distance for axis image 3 and final image 3
hausdorff_filter_3 = sitk.HausdorffDistanceImageFilter()
hausdorff_filter_3.Execute(sitk_initial_image_3, sitk_final_image_3)
hausdorff_sitk_3 = hausdorff_filter_3.GetHausdorffDistance()

```
