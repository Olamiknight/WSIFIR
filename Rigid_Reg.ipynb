```python
#May need to rerun this cell to get the correct results

#Convert the cropped images to ants

axis_image_cropped_ants_1 = ants.from_numpy(cropped_images[f'sampleid_{results[0]["axis_sampleid"]}_image'].compute().data)
tbet_image_cropped_ants_1 = ants.from_numpy(cropped_images[f'sampleid_{results[0]["tbet_sampleid"]}_image'].compute().data)

axis_image_cropped_ants_2 = ants.from_numpy(cropped_images[f'sampleid_{results[1]["axis_sampleid"]}_image'].compute().data)
tbet_image_cropped_ants_2 = ants.from_numpy(cropped_images[f'sampleid_{results[1]["tbet_sampleid"]}_image'].compute().data)

axis_image_cropped_ants_3 = ants.from_numpy(cropped_images[f'sampleid_{results[2]["axis_sampleid"]}_image'].compute().data)
tbet_image_cropped_ants_3 = ants.from_numpy(cropped_images[f'sampleid_{results[2]["tbet_sampleid"]}_image'].compute().data)

# Perform rigid registration first on the first image
rigid_result_1 = ants.registration(
    fixed=axis_image_cropped_ants_1,
    moving=tbet_image_cropped_ants_1,
    type_of_transform='Rigid',
    verbose=True,
)

# Apply the rigid transformation to the first moving image
rigid_transformed_image_1 = ants.apply_transforms(
    fixed=axis_image_cropped_ants_1,
    moving=tbet_image_cropped_ants_1,
    transformlist=rigid_result_1['fwdtransforms']
)



# Perform rigid registration first on the second image
rigid_result_2 = ants.registration(
    fixed=axis_image_cropped_ants_2,
    moving=tbet_image_cropped_ants_2,
    type_of_transform='Rigid',
    verbose=True,
)

# Apply the rigid transformation to the second moving image
rigid_transformed_image_2 = ants.apply_transforms(
    fixed=axis_image_cropped_ants_2,
    moving=tbet_image_cropped_ants_2,
    transformlist=rigid_result_2['fwdtransforms']
)


# Perform rigid registration first on the third image
rigid_result_3 = ants.registration(
    fixed=axis_image_cropped_ants_3,
    moving=tbet_image_cropped_ants_3,
    type_of_transform='Rigid',
    verbose=True,
)

# Apply the rigid transformation to the third moving image
rigid_transformed_image_3 = ants.apply_transforms(
    fixed=axis_image_cropped_ants_3,
    moving=tbet_image_cropped_ants_3,
    transformlist=rigid_result_3['fwdtransforms']
)

```

```python
#Convert the cropped images to ants
  
reg_image_cropped_ants_1 = ants.from_numpy(cropped_images[f'sampleid_{results[0]["reg_sampleid"]}_image'].compute().data)
reg_image_cropped_ants_2 = ants.from_numpy(cropped_images[f'sampleid_{results[1]["reg_sampleid"]}_image'].compute().data)
reg_image_cropped_ants_3 = ants.from_numpy(cropped_images[f'sampleid_{results[2]["reg_sampleid"]}_image'].compute().data)

#Perform rigid registration on the first regression image
final_result_1 = ants.registration(
    fixed=rigid_transformed_image_1,
    moving=reg_image_cropped_ants_1,
    type_of_transform='Rigid',
    verbose=True,
    outprefix='C:\\Users\\molanre1\\Desktop\\WSIFIR\\rigid_registration_result_1'
)
# Apply the rigid transformation to the first regression image
final_image_1 = ants.apply_transforms(
    fixed=rigid_transformed_image_1,
    moving=reg_image_cropped_ants_1,
    transformlist=final_result_1['fwdtransforms']
    
)

#Perform rigid registration on the second regression image
final_result_2 = ants.registration(
    fixed=rigid_transformed_image_2,
    moving=reg_image_cropped_ants_2,
    type_of_transform='Rigid',
    verbose=True,
    outprefix='C:\\Users\\molanre1\\Desktop\\WSIFIR\\rigid_registration_result_2'
)

# Apply the rigid transformation to the second regression image
final_image_2 = ants.apply_transforms(
    fixed=rigid_transformed_image_2,
    moving=reg_image_cropped_ants_2,
    transformlist=final_result_2['fwdtransforms']
    
)

#Perform rigid registration on the third regression image
final_result_3 = ants.registration(
    fixed=rigid_transformed_image_3,
    moving=reg_image_cropped_ants_3,
    type_of_transform='Rigid',
    verbose=True,
    outprefix='C:\\Users\\molanre1\\Desktop\\WSIFIR\rigid_registration_result_3'
)

# Apply the rigid transformation to the third regression image
final_image_3 = ants.apply_transforms(
    fixed=rigid_transformed_image_3,
    moving=reg_image_cropped_ants_3,
    transformlist=final_result_3['fwdtransforms']
    
)


```

```python


# Get the shapes of the images
reg_image_cropped_shape_1 = reg_image_cropped_ants_1.shape
rigid_transformed_image_1_shape = rigid_transformed_image_1.shape

reg_image_cropped_shape_2 = reg_image_cropped_ants_2.shape
rigid_transformed_image_2_shape = rigid_transformed_image_2.shape

reg_image_cropped_shape_3 = reg_image_cropped_ants_3.shape
rigid_transformed_image_3_shape = rigid_transformed_image_3.shape


#Get the target shapes for padding
target_shape_1_x = max(reg_image_cropped_shape_1[0],rigid_transformed_image_1_shape[0])
target_shape_1_y = max(reg_image_cropped_shape_1[1],rigid_transformed_image_1_shape[1])

target_shape_2_x = max(reg_image_cropped_shape_2[0],rigid_transformed_image_2_shape[0])
target_shape_2_y = max(reg_image_cropped_shape_2[1],rigid_transformed_image_2_shape[1])

target_shape_3_x = max(reg_image_cropped_shape_3[0],rigid_transformed_image_3_shape[0])
target_shape_3_y = max(reg_image_cropped_shape_3[1],rigid_transformed_image_3_shape[1])

#Calculate the padding
pad_reg_X_1 = (0,target_shape_1_x-reg_image_cropped_shape_1[0])
pad_reg_Y_1 = (0,target_shape_1_y-reg_image_cropped_shape_1[1])

pad_reg_X_2 = (0,target_shape_2_x-reg_image_cropped_shape_2[0])
pad_reg_Y_2 = (0,target_shape_2_y-reg_image_cropped_shape_2[1])

pad_reg_X_3 = (0,target_shape_3_x-reg_image_cropped_shape_3[0])
pad_reg_Y_3 = (0,target_shape_3_y-reg_image_cropped_shape_3[1])



# Pad the cropped images to match the size of the rigid transformed images
padded_reg_image_ants_1 = ants.pad_image(reg_image_cropped_ants_1, (target_shape_1_x,target_shape_1_y))
padded_reg_image_ants_2 = ants.pad_image(reg_image_cropped_ants_2, (target_shape_2_x,target_shape_2_y))
padded_reg_image_ants_3 = ants.pad_image(reg_image_cropped_ants_3, (target_shape_3_x,target_shape_3_y))


# Pad the axis images to match the size of the rigid transformed images
axis_image_cropped_ants_1_padded = ants.pad_image(axis_image_cropped_ants_1, (target_shape_1_x,target_shape_1_y))
axis_image_cropped_ants_2_padded = ants.pad_image(axis_image_cropped_ants_2, (target_shape_2_x,target_shape_2_y))
axis_image_cropped_ants_3_padded = ants.pad_image(axis_image_cropped_ants_3, (target_shape_3_x,target_shape_3_y))

# Pad the rigid transformed images to match the size of the registered images

rigid_transformed_image_1_padded = ants.pad_image(rigid_transformed_image_1, (target_shape_1_x,target_shape_1_y))
rigid_transformed_image_2_padded = ants.pad_image(rigid_transformed_image_2, (target_shape_2_x,target_shape_2_y))
rigid_transformed_image_3_padded = ants.pad_image(rigid_transformed_image_3, (target_shape_3_x,target_shape_3_y))


#Pad the tbet images to match the size of the registered images
tbet_image_cropped_ants_1_padded = ants.pad_image(tbet_image_cropped_ants_1, (target_shape_1_x,target_shape_1_y))
tbet_image_cropped_ants_2_padded = ants.pad_image(tbet_image_cropped_ants_2, (target_shape_2_x,target_shape_2_y))
tbet_image_cropped_ants_3_padded = ants.pad_image(tbet_image_cropped_ants_3, (target_shape_3_x,target_shape_3_y))


#Pad the final images to match the size of the registered images
final_image_1_padded = ants.pad_image(final_image_1, (target_shape_1_x,target_shape_1_y))
final_image_2_padded = ants.pad_image(final_image_2, (target_shape_2_x,target_shape_2_y))
final_image_3_padded = ants.pad_image(final_image_3, (target_shape_3_x,target_shape_3_y))



# Print the shapes of the padded images
print("Padded axis image shape 1: ", axis_image_cropped_ants_1_padded.shape)
print("Padded axis image shape 2: ", axis_image_cropped_ants_2_padded.shape)
print("Padded axis image shape 3: ", axis_image_cropped_ants_3_padded.shape)


print("Padded regression image shape 1: ", padded_reg_image_ants_1.shape)
print("Padded regression image shape 2: ", padded_reg_image_ants_2.shape)
print("Padded regression image shape 3: ", padded_reg_image_ants_3.shape)

print("Padded tbet image shape 1: ", tbet_image_cropped_ants_1_padded.shape)
print("Padded tbet image shape 2: ", tbet_image_cropped_ants_2_padded.shape)
print("Padded tbet image shape 3: ", tbet_image_cropped_ants_3_padded.shape)

print("Padded rigid transformed image shape 1: ", rigid_transformed_image_1_padded.shape)
print("Padded rigid transformed image shape 2: ", rigid_transformed_image_2_padded.shape)
print("Padded rigid transformed image shape 3: ", rigid_transformed_image_3_padded.shape)

print("Padded final image shape 1: ", final_image_1_padded.shape)
print("Padded final image shape 2: ", final_image_2_padded.shape)
print("Padded final image shape 3: ", final_image_3_padded.shape)

```
